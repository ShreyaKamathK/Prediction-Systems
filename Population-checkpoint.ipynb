```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Sample DataFrame - Replace this with your actual DataFrame (df)
data = {
    'ID': [1, 2, 3, 4, 5],
    'Age': [25, 45, 39, 35, 35],
    'Experience': [1, 19, 15, 9, 8],
    'Income': [49, 34, 11, 100, 45],
    'ZIP Code': [91107, 90089, 94720, 94112, 91330],
    'Family': [4, 3, 1, 1, 4],
    'CCAvg': ['1/60', '1/50', '1/00', '2/70', '1/00'],
    'Education': [1, 1, 1, 2, 2],
    'Mortgage': [0, 0, 0, 0, 0],
    'Personal Loan': [0, 0, 0, 0, 0],
    'Securities Account': [1, 1, 0, 0, 0],
    'CD Account': [0, 0, 0, 0, 0],
    'Online': [0, 0, 0, 0, 0],
    'CreditCard': [0, 0, 0, 0, 1]
}

# Create a DataFrame from the sample data
df = pd.DataFrame(data)

# Assuming 'Personal Loan' column is the target variable (y)
y = df['Personal Loan']
X = df.drop('Personal Loan', axis=1)  # Assuming 'Personal Loan' is the target variable

# Select columns to scale
columns_to_scale = ['Age', 'Experience', 'Income', 'Family']  # Update this list as needed

# Splitting the data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initializing the StandardScaler
scaler = StandardScaler()

# Fitting and transforming the specified columns in the training set
X_train_scaled = X_train.copy()  # Create a copy of the train set to apply scaling
X_train_scaled[columns_to_scale] = scaler.fit_transform(X_train[columns_to_scale])

# Transforming the same columns in the test set (using the parameters learned from the training set)
X_test_scaled = X_test.copy()  # Create a copy of the test set to apply scaling
X_test_scaled[columns_to_scale] = scaler.transform(X_test[columns_to_scale])

# Print the modified datasets (scaled columns)
print("Scaled X_train:")
print(X_train_scaled.head())
print("\nScaled X_test:")
print(X_test_scaled.head())
```
